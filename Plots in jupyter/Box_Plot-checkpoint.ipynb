```python
import seaborn as sns


kashti=sns.load_dataset('titanic')
sns.boxplot(x='survived', y='age', data=kashti)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

kashti=sns.load_dataset('titanic')
sns.boxplot(x='survived', y='age', showmeans=True,meanprops={'marker':'*',
                                                            'markersize':'40',
                                                            'markeredgecolor':'blue'},
            data=kashti)
# Show Labels
plt.xlabel('How many of u survived')
plt.ylabel('Survived age')
plt.title('Graph of Survived Guys')
```

# Facet_wrap and Facet_grid

```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset('tips')
garph=sns.FacetGrid(df, col='sex', hue='day')
garph.map(plt.scatter, 'total_bill', 'tip', edgecolor='blue')
plt.show()
```

```python

# importing packages
import seaborn
import matplotlib.pyplot as plt
  
# loading of a dataframe from seaborn
df = seaborn.load_dataset('tips')
  
############# Main Section         #############
# Form a facetgrid using columns with a hue
graph = seaborn.FacetGrid(df, row ='smoker', col ='time')
# map the above form facetgrid with some attributes
graph.map(plt.hist, 'total_bill', bins = 15, color ='orange')
# show the object
plt.show()
```
